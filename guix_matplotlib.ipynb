```guile
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```guile
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)

plt.title("test", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.ylabel('X Axis', fontdict={'fontname': 'Ariel', 'fontsize': 40})

plt.show()
```

```guile
x = [0,1,2,3,4,5,6]
y = [0,2,4,6,8,10,12]

plt.plot(x,y, label="2x", color="red", linewidth=5,linestyle='--',marker='.',markersize=10,markeredgecolor='blue')

plt.title("test" )
plt.ylabel('X Axis')

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

```guile
x = [0,1,2,3,4,5,6]
y = [0,2,4,6,8,10,12]

# Use shorthand notion
# fmt = '[color][marker][line]'
plt.plot(x,y, 'g^--', label="2x")

plt.title("test" )
plt.ylabel('X Axis')

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

```guile
x = [0,1,2,3,4,5,6]
y = [0,2,4,6,8,10,12]

# Use shorthand notion
# fmt = '[color][marker][line]'
plt.plot(x,y, 'g^--', label="2x")

## Line 2
# [0.  0.5 1.  1.5 2.  2.5 3.  3.5]
x2 = np.arange(0,4,0.5)
plt.plot(x2, x2**2, 'bp:', label="x^2")

plt.title("test" )
plt.ylabel('X Axis')

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

# Bar Charts


```guile
labels = ['a', 'b', 'c']
values = [1,4,2]

plt.bar(labels, values)

plt.show()
```

```guile
labels = ['a', 'b', 'c']
values = [1,4,2]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')

plt.show()
```
